```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_daily_solcast = pd.read_csv("solcast/solcast_nov_2025_daily_avg.csv", parse_dates=["date"])

df_daily_solcast = df_daily_solcast.rename(columns={"date": "Date", "ghi_avg": "GHI"})



df_daily_kbg = pd.read_csv("Actual Sensor Data_non2025.csv", sep=";", header=0, names=["Date", "GHI"])

df_daily_kbg["Date"] = pd.to_datetime(df_daily_kbg["Date"], format="%b %d %Y")



# Open-Meteo daily solar (kebithigollewa); using Direct Normal Irradiance (dni_avg) in W/mÂ²

df_openmeteo = pd.read_csv("openmeteo/kebithigollewa_openmeteo_solar_daily_2025_11.csv", parse_dates=["date"])

df_openmeteo = df_openmeteo.rename(columns={"date": "Date", "ghi_avg": "Direct_Wm2"})
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_daily_solcast["Date"], df_daily_solcast["GHI"], marker='o', label='Solcast GHI', color='orange')
plt.plot(df_daily_kbg["Date"], df_daily_kbg["GHI"], marker='x', label='KBG GHI', color='blue')
plt.plot(df_openmeteo["Date"], df_openmeteo["Direct_Wm2"], marker='s', label='Open-Meteo Direct (avg W/mÂ²)', color='green')
plt.xlabel('Date')
plt.ylabel('Irradiance (W/mÂ²)')
plt.title('Daily Average Irradiance - November 2025')
plt.legend()
plt.grid()
plt.show()
```

```python
df_daily_solcast = pd.read_csv("solcast/solcast_nov_2025_daily_avg.csv", parse_dates=["date"])

df_daily_solcast = df_daily_solcast.rename(columns={"date": "Date", "ghi_avg": "GHI"})



df_daily_kbg = pd.read_csv("Actual Sensor Data_non2025.csv", sep=";", header=0, names=["Date", "GHI"])

df_daily_kbg["Date"] = pd.to_datetime(df_daily_kbg["Date"], format="%b %d %Y")



# Open-Meteo daily solar (kebithigollewa); using Direct Normal Irradiance (dni_avg) in W/mÂ²

df_openmeteo = pd.read_csv("kebithigollewa_openmeteo_solar_daily_api_2025_11.csv", parse_dates=["time"])

df_openmeteo = df_openmeteo.rename(columns={"time": "Date", "shortwave_radiation": "Direct_Wm2"})
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_daily_solcast["Date"], df_daily_solcast["GHI"], marker='o', label='Solcast GHI', color='orange')
plt.plot(df_daily_kbg["Date"], df_daily_kbg["GHI"], marker='x', label='KBG GHI', color='blue')
plt.plot(df_openmeteo["Date"], df_openmeteo["Direct_Wm2"]/12, marker='s', label='Open-Meteo Direct (avg W/mÂ²)', color='green')
plt.xlabel('Date')
plt.ylabel('Irradiance (W/mÂ²)')
plt.title('Daily Average Irradiance - November 2025')
plt.legend()
plt.grid()
plt.show()
```

```python
# Compare Open-Meteo daily (processed) vs API download
df_openmeteo_daily = pd.read_csv("openmeteo/kebithigollewa_openmeteo_solar_daily_2025_11.csv", parse_dates=["date"])
df_openmeteo_daily = df_openmeteo_daily.rename(columns={"date": "Date", "ghi_avg": "Direct_Wm2"})

df_openmeteo_api = pd.read_csv("kebithigollewa_openmeteo_solar_daily_api_2025_11.csv", parse_dates=["time"])
df_openmeteo_api = df_openmeteo_api.rename(columns={"time": "Date", "shortwave_radiation": "Direct_Wm2"})

plt.figure(figsize=(12, 6))
plt.plot(df_openmeteo_daily["Date"], df_openmeteo_daily["Direct_Wm2"], marker='o', label='Open-Meteo processed (GHI avg)', color='green')
plt.plot(df_openmeteo_api["Date"], df_openmeteo_api["Direct_Wm2"]/12, marker='s', label='Open-Meteo API shortwave', color='purple')
plt.xlabel('Date')
plt.ylabel('Irradiance (W/mÂ²)')
plt.title('Open-Meteo Daily Irradiance - November 2025')
plt.legend()
plt.grid()
plt.show()
```
